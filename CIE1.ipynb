```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/mtcars (1).csv')
data
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['mpg'], bins=10, edgecolor='k', color='skyblue')
plt.title('Histogram of MPG (Miles per gallon)')
plt.xlabel('MPG')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['wt'], data['mpg'], color='red', marker='^')
plt.title('Scatter Plot: Car Weight vs. MPG')
plt.xlabel('Car Weight')
plt.ylabel('MPG')

plt.show()
```

```python
transmission_counts = data['am'].value_counts()
plt.figure(figsize=(8, 6))
transmission_counts.plot(kind='bar', color='red')
plt.title('Frequency Distribution of Transmission Type')
plt.xlabel('Transmission Type (0 = Automatic, 1 = Manual)')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data['mpg'], patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box and Whisker Plot of MPG')
plt.xlabel('MPG')
plt.show()
```
